```python
from tkinter import *
import numpy as np
from PIL import Image, ImageTk
import tkinter.messagebox as msgbox
import pandas as pd
import time
import os


root = Tk()
root.title('Black jack')
root.geometry("1200x600")
root.configure(background="green")

# creating my folder directory(if you want to open you have to copy the link of the folder)

my_directory = r'C:\Users\vila0\Desktop\BJ project/'

#main frames

cards_frame = Frame(root, bg= 'green')
cards_frame.grid(row=0, padx = 10, pady = 10)


labels_frame = Frame(root, bg="green")
labels_frame.grid(row=2, padx = 10, pady = 10)

button_frame = Frame(root, bg="green")
button_frame.grid(row=1, padx = 10, pady = 10)



#closed card label

closed_card_jpg =ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/uzversta_korta.jpg").resize((100, 150))) 

# computer frame and first cards labels

comp_card1_2_frame = LabelFrame(cards_frame, bd=0, bg='green')
comp_card1_2_frame.grid(row=1, column=1)

comp_card1 = Label(comp_card1_2_frame,image=closed_card_jpg, bg = "green")
comp_card1.grid(row=0,column=0)

comp_card2 = Label(comp_card1_2_frame,image=closed_card_jpg, bg = "green")
comp_card2.grid(row=0,column=1)

# player frame and first cards labels


player_card1_2_frame = LabelFrame(cards_frame, bd=0, bg='green')
player_card1_2_frame.grid(row=4, column=1)

player_card1 = Label(player_card1_2_frame,image=closed_card_jpg, bg = "green")
player_card1.grid(row=0,column=0)

player_card2 = Label(player_card1_2_frame,image=closed_card_jpg, bg = "green")
player_card2.grid(row=0,column=1)

#labels

comp_score = Label(cards_frame, text = 0, font = 100, bg = "grey", fg = "black")
comp_score.grid(row=1, column=0, padx = 10, pady = 10)

player_score_label = Label(cards_frame, text = 0, font = 100, bg = "grey", fg = "black")
player_score_label.grid(row=4, column=0, padx = 10, pady = 10)

user_indicator = Label(cards_frame,font = 50, text = "Player")
computer_indicator = Label(cards_frame,font = 50, text = "Computer" )

user_indicator.grid(row=3,column=1, padx = 10, pady = 10) 
computer_indicator.grid(row=0,column=1, padx = 10, pady = 10)

#text label

text_label = Label(labels_frame,font =150, text = 'Welcome to  BLACK JACK,\n please choose you bid', bg = "green", fg ="Black",)
text_label.grid(row = 1, column =5, padx = 10, pady = 10)

# CREATING cards open pictures dictionary

cards_dict = {}

for card_index in range(2,12):
    if card_index in range(2,11):
        cards_dict[f'{card_index}_sirdys'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/{card_index}_sirdys.jpg").resize((100, 150))),
        cards_dict[f'{card_index}_vynai'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/{card_index}_vynai.jpg").resize((100, 150))),
        cards_dict[f'{card_index}_bugnai'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/{card_index}_bugnai.jpg").resize((100, 150))),
        cards_dict[f'{card_index}_kryziai'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/{card_index}_kryziai.jpg").resize((100, 150)))
    else:
        cards_dict['J_sirdys'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/J_sirdys.jpg").resize((100, 150)))
        cards_dict['Q_sirdys'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/Q_sirdys.jpg").resize((100, 150)))
        cards_dict['K_sirdys'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/K_sirdys.jpg").resize((100, 150)))
        cards_dict['A_sirdys'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/A_sirdys.jpg").resize((100, 150)))

        cards_dict['J_vynai'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/J_vynai.jpg").resize((100, 150)))
        cards_dict['Q_vynai'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/Q_vynai.jpg").resize((100, 150)))
        cards_dict['K_vynai'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/K_vynai.jpg").resize((100, 150)))
        cards_dict['A_vynai'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/A_vynai.jpg").resize((100, 150)))

        cards_dict['J_bugnai'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/J_bugnai.jpg").resize((100, 150)))
        cards_dict['Q_bugnai'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/Q_bugnai.jpg").resize((100, 150)))
        cards_dict['K_bugnai'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/K_bugnai.jpg").resize((100, 150)))
        cards_dict['A_bugnai'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/A_bugnai.jpg").resize((100, 150)))

        cards_dict['J_kryziai'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/J_kryziai.jpg").resize((100, 150)))
        cards_dict['Q_kryziai'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/Q_kryziai.jpg").resize((100, 150)))
        cards_dict['K_kryziai'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/K_kryziai.jpg").resize((100, 150)))
        cards_dict['A_kryziai'] = ImageTk.PhotoImage(Image.open(f"{my_directory}/cards/A_kryziai.jpg").resize((100, 150)))


# cards_dic ---> list x3

cards_list = (list(cards_dict) *3)

#empty variables

bid = 0

game_status = ''

#player class

class Player_class:
    def __init__(self, player_name, player_bank):
        self.player_name = player_name
        self.player_bank = player_bank
    def bet(self, bid):
        text_label['text'] = f'Your bid is {bid}€,\n press "OPEN CARDS "to start the game'
        return bid

#create player within class

my_player = Player_class(player_bank=100, player_name= "Vilgirdas")    

#bank label

bank = Label(cards_frame,font = 100, text =((f"Player's {my_player.player_name} bank is {my_player.player_bank}€")))
bank.grid(row=0,column=15) 

#Empty dataframe

game_data_frame = pd.DataFrame({'player_cards': [],
'computer_cards': [],
'player_points': [],
'computer_points':[],
'winner':[],
'player_bid': [],
'bank': []
})

#Function which concating dataframes

def upload_data():
    global game_data_frame
    new_row = pd.DataFrame({'player_cards': [', '.join(player_hand_cards_list)],'computer_cards': [', '.join(comp_hand_cards_list)],'player_points': [hand_points(player_hand_cards_list)],'computer_points':[hand_points(comp_hand_cards_list)],'winner': [game_status], 'player_bid': [bid],'bank': [my_player.player_bank]})
    game_data_frame = pd.concat([game_data_frame, new_row], axis = 0)
    return game_data_frame

#Functions

#takes card from list

def card_taking(card_list):
  card = np.random.choice(card_list)
  card_list.remove(card)
  return card

#deal_cards for one hand

def deal_cards():
    hand_card = [card_taking(cards_list), card_taking(cards_list)]
    return hand_card

#calc hand card points

def hand_points(hand_cards):
    total = 0
    A_card = 0
    for card in hand_cards:
        if card[0] in face or card[:2] in face:
            total +=10
        elif card[0] == "A":
            A_card+=1
            total += 11
        elif int(card[0]) in range(2,10):
            total+=int(card[0])
    while (total >21) and (A_card > 0):
        A_card-=1
        total = total -(11-1)    
    return total

#OPENS player cards and first computer card

def open_first_cards(player_cards, comp_cards, label1_player, label2_player, label1_comp, dic):
    label1_player.configure(image = dic[player_cards[0]])
    label2_player.configure(image = dic[player_cards[1]])
    label1_comp.configure(image = dic[comp_cards[0]])
    text_label['text'] = f'If you think that {hand_points(player_hand_cards_list)} is enough, press "Stand",\n if you want to risk and take extra card press "HIT"'
    if (comp_cards[0][0] in face ) or (comp_cards[0][:2] in face):
        comp_score["text"] = 10
    elif comp_cards[0][0] == "A":
        comp_score["text"] = 11
    else: 
        comp_score["text"] = comp_cards[0][0]
    update_player_points_label()
    update_bank(my_player.player_bank)

    STAND_button.grid(row=4, column = 0)
    HIT_button.grid(row=1, column = 0)
    OPEN_CARDS_button.grid_forget()

    if hand_points(player_hand_cards_list) == 21:
        computer_points_calc()
        find_winner(player_total_score = hand_points(player_hand_cards_list), comp_total_score=computer_points_calc())

#Update bank label on screen

def update_bank(x):
    global my_player
    bank["text"]  = (f"Player's {my_player.player_name} bank is {(str(x))}€")

#Computer cards taking until over 17 and retruns points

def computer_points_calc(dic = cards_dict):
    global comp_hand_cards_list
    global comp_card_index
    comp_card2['image'] = dic[comp_hand_cards_list[1]]
    comp_score["text"]=str(hand_points(comp_hand_cards_list))
    if hand_points(comp_hand_cards_list) >= 18:
        return hand_points(comp_hand_cards_list)
    else:
        while hand_points(comp_hand_cards_list)< 18:
            dealer_extra_card = card_taking(cards_list)
            comp_hand_cards_list.append(dealer_extra_card)
            comp_card_index+=1
            globals()[f'comp_card{comp_card_index}_frame'] = LabelFrame(cards_frame, bd=0, bg='green')
            globals()[f'comp_card{comp_card_index}_frame'].grid(row=1, column=(comp_card_index-1))
            globals()[f'comp_card{comp_card_index}'] = Label(globals()[f'comp_card{comp_card_index}_frame'],image= dic[dealer_extra_card], bg = "green")
            globals()[f'comp_card{comp_card_index}'].pack()
            if hand_points(comp_hand_cards_list) >= 18:
                comp_score["text"]=str(hand_points(comp_hand_cards_list))
                return hand_points(comp_hand_cards_list)

#Player extra card taking 

def player_hit(card_list, dic):
    global player_card_index
    global player_hand_cards_list
    player_card_index+=1
    extra_card = card_taking(card_list)
    player_hand_cards_list.append(extra_card)
    globals()[f'player_card{player_card_index}_frame'] = LabelFrame(cards_frame, bd=0, bg='green')
    globals()[f'player_card{player_card_index}_frame'].grid(row=4, column=(player_card_index-1))
    globals()[f'player_card{player_card_index}'] = Label(globals()[f'player_card{player_card_index}_frame'],image= dic[extra_card], bg = "green")
    globals()[f'player_card{player_card_index}'].pack()

    if hand_points(player_hand_cards_list) >=21:
        find_winner(player_total_score = hand_points(player_hand_cards_list), comp_total_score=computer_points_calc())            
    else:
        text_label['text'] = f'If you think that {hand_points(player_hand_cards_list)} is enough, press "Stand",\n if you want to risk and take extra card press "HIT"'

    update_player_points_label()

#update player score on screen

def update_player_points_label():
    points = int(player_score_label["text"])
    points = hand_points(player_hand_cards_list)
    player_score_label["text"]=str(points)

#text label update on screen

def update_text_label(x):
    text_label['text'] = x

#Points comparing

def find_winner (player_total_score, comp_total_score):
    global game_status
    if player_total_score == 21:
        if player_total_score == comp_total_score:
            update_text_label(f"It is DRAW! +{bid}€!")
            my_player.player_bank+=bid
            game_status = "DRAW"
        else:
             my_player.player_bank+=(bid*3)
             update_text_label(f"YOU GOT BLACK JACK!(21) +{bid *3}€")
             game_status = "WON"
    else:
        if player_total_score > 21:
            update_text_label(f"YOU LOST -{bid}€")
            game_status = "LOST"

        elif player_total_score == comp_total_score:
            my_player.player_bank+=bid
            update_text_label(f"It is DRAW! +{bid}€!")
            game_status = "DRAW"
        elif player_total_score < comp_total_score and comp_total_score < 22:
            update_text_label(f"YOU LOST -{bid}€")
            game_status = "LOST"
        elif player_total_score > comp_total_score or comp_total_score > 21:
            my_player.player_bank+=(bid *2)
            update_text_label(f"Congrats you WON! +{bid *2}€")
            game_status = "WON"
    update_player_points_label()
    update_bank(my_player.player_bank)
    HIT_button.grid_forget()
    STAND_button.grid_forget()
    END_GAME_button.grid(row=1, column = 6)
    PLAY_AGAIN_button.grid(row=1, column = 5)
    upload_data()

#playing again

def play_again():
    global player_hand_cards_list
    global comp_hand_cards_list
    global player_card_index
    global comp_card_index


    for sk in range(1,(player_card_index+1)):
        if sk in range(1,3):
            (globals()[f'player_card{sk}'])['image'] =closed_card_jpg
        else:
            (globals()[f'player_card{sk}'])['image'] =''

    for sk in range(1,(comp_card_index+1)):
        if sk in range(1,3):
            (globals()[f'comp_card{sk}'])['image'] = closed_card_jpg
        else:
            (globals()[f'comp_card{sk}'])['image'] =''

    player_hand_cards_list = []
    comp_hand_cards_list = []

    player_hand_cards_list = deal_cards()
    comp_hand_cards_list = deal_cards()

    player_card_index = 2
    comp_card_index = 2

    text_label['text'] = ''
    comp_score ['text']  = 0
    player_score_label['text']  = 0

    PLAY_AGAIN_button.grid_forget()
    END_GAME_button.grid_forget()

    if my_player.player_bank <5:
        update_text_label("Your bank is empty, can not continue playing")
        END_GAME_button.grid(row=1, column = 2)
    elif my_player.player_bank <10:
        BID5_button.grid(row=1, column = 2)
        ALL_IN_button.grid(row=1, column = 4)
    else:
        BID5_button.grid(row=1, column = 2)
        BID10_button.grid(row=1, column = 3)
        ALL_IN_button.grid(row=1, column = 4)

# bidding

def betting(suma):
    global bid
    bid = my_player.bet(suma)
    my_player.player_bank-= bid
    update_bank(my_player.player_bank)
    OPEN_CARDS_button.grid(row=1, column = 0)
    BID10_button.grid_forget()
    BID5_button.grid_forget()
    ALL_IN_button.grid_forget()

 #GAME ending and request for dataframe

def end_game():
    END_GAME_button.grid_forget()
    PLAY_AGAIN_button.grid_forget()
    YES_button.grid(row=1, column = 0)
    NO_button.grid(row=1, column = 1)
    update_text_label("Do you want to save results?")

# Datafrane to excel by name of time, ends the game

def yes():
    global game_data_frame
    print(game_data_frame)
    END_GAME_button.grid_forget()

    timestr = time.strftime("%Y%m%d-%H%M")
    print (timestr)
    file_name = f'Black_Jack_game{timestr}.xlsx'
    game_data_frame.to_excel(file_name, index = False)

    msgbox.showinfo(title='', message= f'Thank you!, your results saved into file named {file_name}')
    root.destroy()


#GAME

# Dealing cards

player_hand_cards_list = deal_cards()
comp_hand_cards_list = deal_cards()

# Faces and card with 10 points

face = ['J', 'K', 'Q', '10']

# cards index on table

player_card_index = 2
comp_card_index = 2

#Buttons

HIT_button = Button(button_frame,width = 20, height = 2, text = "HIT", bg = "black", fg = "red",state =NORMAL, command = lambda: player_hit(card_list=cards_list, dic = cards_dict))

OPEN_CARDS_button= Button(button_frame,width = 20, height = 2, text = "OPEN CARDS", bg = "black", fg = "red", state =NORMAL, command = lambda: open_first_cards(player_cards = player_hand_cards_list, comp_cards = comp_hand_cards_list, label1_player = player_card1, label2_player =  player_card2, label1_comp = comp_card1, dic =cards_dict))

PLAY_AGAIN_button = Button(button_frame,width = 20, height = 2, text = "PLAY AGAIN", bg = "black", fg = "red", state =NORMAL, command = lambda: play_again())

STAND_button = Button(button_frame,width = 20, height = 2, text = "STAND", bg = "black", fg = "red", state = NORMAL, command=lambda: find_winner (player_total_score = hand_points(player_hand_cards_list), comp_total_score = computer_points_calc()))

BID5_button = Button(button_frame,width = 20, height = 2, text = "5€ ", bg = "black", fg = "red", state = NORMAL, command=lambda: betting(5))
BID5_button.grid(row=1, column = 2)

BID10_button = Button(button_frame,width = 20, height = 2, text = "10€ ", bg = "black", fg = "red", state = NORMAL, command=lambda: betting(10))
BID10_button.grid(row=1, column = 3)

ALL_IN_button = Button(button_frame,width = 20, height = 2, text = f"ALL_IN", bg = "black", fg = "red", state = NORMAL, command=lambda: betting(my_player.player_bank))
ALL_IN_button.grid(row=1, column = 4)

END_GAME_button = Button(button_frame,width = 20, height = 2, text = "END GAME ", bg = "black", fg = "red", state = NORMAL, command= lambda: end_game())

YES_button = Button(button_frame,width = 20, height = 2, text = "YES ", bg = "black", fg = "red", state = NORMAL, command= lambda: yes())

NO_button = Button(button_frame,width = 20, height = 2, text = "NO ", bg = "black", fg = "red", state = NORMAL, command= root.destroy)

#OPEN SCREEN

root.mainloop()






```
